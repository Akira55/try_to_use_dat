```python
import pandas as pd
import glob
from functools import reduce
import re
import os
```

```python
def make_data_country(file_path):

    with open(file_path, 'r') as f:
        total_list = f.read().splitlines()
    
    file_name = file_path.split('.')[0].split('/')[1]
    
    #なければディレクトリをつくる
    dir_path = "country_data/" + file_name
    if not os.path.exists(dir_path):
        os.mkdir(dir_path)    
    conuntry_index = {}
    last_ind = 0
    # 国ごとにdatファイルを分割する
    for index, row in enumerate(total_list):
        flag = re.findall(r"[A-Z]+" , row) #国名だけを抜く
        if len(flag) != 0:
            name = flag[0]
            conuntry_index[name] = [last_ind, index]
            last_ind = index
    
    # 国ごとにdatファイルを保存
    for name in conuntry_index.keys():
        first, last = conuntry_index[name]
        data = total_list[first+1:last]
        with open('country_data/' + file_name + '/' + name + '.dat', mode='w') as f:
            f.write('\n'.join(data))
```

```python
# 国ごとにファイルを作り、もとのdatファイルごとにディレクトリに入れる
for f_path in glob.glob('dat_file/*'):
    make_data_country(f_path)
```

```python

df_list = []
#テストでgu23wld0098を読み込む
for file_path in glob.glob('country_data/gu23wld0098/*.dat'):
    
    par_filename = file_path.split('/')[1]
    country = file_path.split('/')[2].split('.')[0]
    
    # 1-12月と最後の列はhoge
    names = [str(i) for i in range(1, 13)] + ['Hoge']
    # stack 1, 2, ...12, hoge のインデックスを追加するならべる
    # いい感じにdatファイルを読む
    df = pd.read_table(file_path, skiprows=1,  sep='\s+', names=names)\
                        .stack(dropna=False).to_frame()
    # 国名を入れる
    df['country'] = country
    df.reset_index(inplace=True)
    # valuの末尾にファイル名を追加しておく
    df.columns = ['year', 'month', 'value_' + par_filename, 'country']
    df_list.append(df)
```
